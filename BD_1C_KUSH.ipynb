```python
from bs4 import BeautifulSoup
import requests
import pandas as pd
import numpy as np

```

```python
def get_element_text(soup, tag, attrs):
    try:
        element = soup.find(tag, attrs=attrs).string.strip()
    except AttributeError:
        element = ""
    return element
```

```python

def get_title(soup):
    try:
        title = soup.find("span", attrs={"id": 'productTitle'})
        title_value = title.text
        title_string = title_value.strip()
    except AttributeError as e:
        title_string = f"Error: {e}"
    return title_string
```

```python

def get_review_count(soup):
    return get_element_text(soup, "span", {'id': 'acrCustomerReviewText'})
```

```python

def get_rating(soup):
    return get_element_text(soup, "i", {'class': 'a-icon a-icon-star a-star-4-5'}) or get_element_text(soup, "span", {'class': 'a-icon-alt'})

```

```python
def get_availability(soup):
    try:
        available = soup.find("div", attrs={'id': 'availability'})
        available = available.find("span").string.strip()
    except AttributeError:
        available = "Not Available"
    return available
```

```python

def get_price(soup):
    price_element = soup.find("span", attrs={"class": "a-price-whole"})
    if price_element:
        price = price_element.text
        return price
    else:
        return None
```

```python

if __name__ == '__main__':
    HEADERS = {'User-Agent': '', 'Accept-Language': 'en-US, en;q=0.5'}

    URL = 'https://www.amazon.in/s?k=self+help&ref=nb_sb_noss_1'
    webpage = requests.get(URL, headers=HEADERS)
    soup = BeautifulSoup(webpage.content, "html.parser")

    links = soup.find_all("a", attrs={'class': 'a-link-normal s-no-outline'})
    links_list = [link.get('href') for link in links]

    data_list = []

    for link in links_list:
        new_webpage = requests.get("https://www.amazon.in" + link, headers=HEADERS)
        new_soup = BeautifulSoup(new_webpage.content, "html.parser")

        data_list.append({
            'title': get_title(new_soup),
            'price': get_price(new_soup),
            'rating': get_rating(new_soup),
            'reviews': get_review_count(new_soup),
            'availability': get_availability(new_soup)
        })

    df = pd.DataFrame(data_list)


```
